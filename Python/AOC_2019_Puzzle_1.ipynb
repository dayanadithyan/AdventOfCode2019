## AOC 2019 --- Day 1: The Tyranny of the Rocket Equation ---

### Part (1)

```python
import math
```

```python
def fuel_calculator(mass):
    
    fuel_req = math.floor(int(mass) / 3) - 2
    
    return fuel_req
```

```python
fuel_req_list = []
```

```python
with open ('input','r') as input_file:
    lines = input_file.readlines()
    for line in lines:
        f_req = fuel_calculator(line)
        fuel_req_list.append(f_req)
    
    
```

```python
total_fuel = sum(fuel_req_list)
```

```python
total_fuel
```

```python
mod_masses = []

for line in lines:
    mod_masses.append(int(line))
```

```python
mod_masses
```

### Part (2)

```python
fuel_req_list
```

```python
fuel_list_main = []

    
```

```python
%%timeit

for mass in mod_masses:

    fuel_req = fuel_calculator(mass)
    
    fuel_list_main.append(fuel_req)
    
    current_fuel = fuel_req
    
    while (current_fuel /3) - 2 > 2:
        
        add_this = fuel_calculator(current_fuel)
          
        fuel_list_main.append(add_this)

        current_fuel = add_this
        
       
    
    
    
        
        
    
            
        
    
    
    
    
```

```python
fuel_list_main
    
```

```python
new_fuel_requirement = sum(fuel_list_main) 
```
